```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================================
# TF-IDF FEATURE EXTRACTION
# ============================================================================
tfidf = TfidfVectorizer(max_features=5000)
X_train_trans2 = tfidf.fit_transform([" ".join(map(str, x)) for x in X_train["cleaned_text"]])
X_test_trans2 = tfidf.transform([" ".join(map(str, x)) for x in X_test["cleaned_text"]])

print("TF-IDF Shape:")
print(f"X_train_trans2: {X_train_trans2.shape}")
print(f"X_test_trans2: {X_test_trans2.shape}")

# ============================================================================
# LOGISTIC REGRESSION - BAG OF WORDS
# ============================================================================
print("\n" + "="*70)
print("BAG OF WORDS - LOGISTIC REGRESSION")
print("="*70)

model_bow = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
model_bow.fit(X_train_trans2, y_train)
y_pred_bow = model_bow.predict(X_test_trans2)

print(f"Accuracy: {accuracy_score(y_test, y_pred_bow):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_bow, average='weighted'):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_bow, average='weighted'):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_bow, average='weighted'):.4f}")
print("\nClassification Report (BoW):")
print(classification_report(y_test, y_pred_bow))

# ============================================================================
# LOGISTIC REGRESSION - TF-IDF
# ============================================================================
print("\n" + "="*70)
print("TF-IDF - LOGISTIC REGRESSION")
print("="*70)

model_tfidf = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
model_tfidf.fit(X_train_trans2, y_train)
y_pred_tfidf = model_tfidf.predict(X_test_trans2)

print(f"Accuracy: {accuracy_score(y_test, y_pred_tfidf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_tfidf, average='weighted'):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_tfidf, average='weighted'):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_tfidf, average='weighted'):.4f}")
print("\nClassification Report (TF-IDF):")
print(classification_report(y_test, y_pred_tfidf))

# ============================================================================
# COMPARISON
# ============================================================================
print("\n" + "="*70)
print("MODEL COMPARISON")
print("="*70)

comparison = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'BoW': [
        accuracy_score(y_test, y_pred_bow),
        precision_score(y_test, y_pred_bow, average='weighted'),
        recall_score(y_test, y_pred_bow, average='weighted'),
        f1_score(y_test, y_pred_bow, average='weighted')
    ],
    'TF-IDF': [
        accuracy_score(y_test, y_pred_tfidf),
        precision_score(y_test, y_pred_tfidf, average='weighted'),
        recall_score(y_test, y_pred_tfidf, average='weighted'),
        f1_score(y_test, y_pred_tfidf, average='weighted')
    ]
}

import pandas as pd
comparison_df = pd.DataFrame(comparison)
print(comparison_df)

# ============================================================================
# CONFUSION MATRICES
# ============================================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# BoW Confusion Matrix
cm_bow = confusion_matrix(y_test, y_pred_bow)
sns.heatmap(cm_bow, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=True)
axes[0].set_title('Confusion Matrix - BoW')
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')

# TF-IDF Confusion Matrix
cm_tfidf = confusion_matrix(y_test, y_pred_tfidf)
sns.heatmap(cm_tfidf, annot=True, fmt='d', cmap='Greens', ax=axes[1], cbar=True)
axes[1].set_title('Confusion Matrix - TF-IDF')
axes[1].set_ylabel('True Label')
axes[1].set_xlabel('Predicted Label')

plt.tight_layout()
plt.savefig('bow_vs_tfidf_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Models trained and compared successfully!")
```
